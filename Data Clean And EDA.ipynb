```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Admin\Downloads\train.csv")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.drop(columns=['Cabin'], inplace=True)
```

```python
print(data.isnull().sum())
```

```python
sns.countplot(x='Survived', data=data)

plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=data)

plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=data)

plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.hist(data['Age'], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.histplot(data['Fare'], bins=20, kde=True)

plt.title("Fare Distribution")
plt.show()
```

```python
#Final Conclusion

#Using data cleaning and EDA, we identified important patterns

#Gender and passenger class strongly influenced survival.

#Wealthier passengers had better chances of survival.

#Age distribution showed most passengers were young adults.
```
