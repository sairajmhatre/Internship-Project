```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
```

```python
data = pd.read_csv(r"C:\Users\Admin\Downloads\twitter_training.csv" ,encoding='latin1')
```

```python
print(data.head())
```

```python
data.columns = ['ID', 'Topic', 'Sentiment', 'Text']
```

```python
#check data
print(data.info())
```

```python
print(data['Sentiment'].value_counts())
```

```python
#data cleaning 
data = data.dropna()
```

```python
data['Text'] = data['Text'].str.lower()
```

```python
sns.countplot(x='Sentiment', data=data)
```

```python
plt.title("Sentiment Distribution")
plt.show()
```

```python
sns.countplot(x='Sentiment', data=data)

plt.title("Sentiment Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(x='Topic', hue='Sentiment', data=data)

plt.title("Sentiment by Topic")
plt.show()
```
